# Iris Dataset EDA - Internship Task 1

"""
Task 1 involves exploring and visualizing the Iris dataset to understand 
data relationships, distributions, and possible outliers.
"""


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: Set visualization style
sns.set(style="whitegrid")
```

```python
# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Preview the dataset
df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nSummary statistics:")
print(df.describe())
```

# Scatter Plot: Sepal vs Petal

```python
# Scatter plot to explore relationships
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

# Histogram: Distribution of Sepal Length

```python
# Histogram
plt.figure(figsize=(7, 5))
sns.histplot(df['sepal_length'], kde=True, color='skyblue')
plt.title("Distribution of Sepal Length")
plt.show()
```

# Box Plot: Outlier Detection

```python
# Box plot by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="species", y="sepal_width")
plt.title("Boxplot of Sepal Width by Species")
plt.show()
```
